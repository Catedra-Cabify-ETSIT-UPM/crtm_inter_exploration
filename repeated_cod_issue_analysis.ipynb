```python
import pandas as pd # main data processing tool
import datetime # for datetime object opperations
import time # for measuring the processing time
from _plotly_future_ import renderer_defaults, timezones
import plotly.express as px # for graphs
```

```python
df = pd.read_csv(
                     'out_parsed',
                     parse_dates=['actual_date', 'eta'],
                     nrows=1e7
                     )
df['remaining_seconds'] = df['eta'] - df['actual_date']
df['eta_date'] = df['eta'].dt.day
```

```python
plot_df = df.query("cod_line == '8_N_901___' & eta_date == 18 & cod_stop == '8_06277'")
```

```python
fig = px.scatter(plot_df,
                 x=plot_df['actual_date'].dt.to_pydatetime(),
                 y=plot_df['eta'].dt.to_pydatetime(),
                 color='cod_issue')
# Edit the layout
fig.update_layout(title='ETA over time',
                  xaxis_title='Sample collection date',
                  yaxis_title='ETA')
fig.show()
```
