# Visualization of the model performance

For the CRTM API, the Random Forest Regressor and the Model Tree.

```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
```

```python
mae_global_crtm = pd.read_csv('mae_global_crtm.csv')
mae_greens_crtm = pd.read_csv('mae_greens_crtm.csv')
mae_reds_crtm = pd.read_csv('mae_reds_crtm.csv')
```

```python
mae_global_mt = pd.read_csv('mae_global_mt.csv')
mae_greens_mt = pd.read_csv('mae_greens_mt.csv')
mae_reds_mt = pd.read_csv('mae_reds_mt.csv')
```

```python
mae_global_rfr = pd.read_csv('mae_global_rfr.csv')
mae_greens_rfr = pd.read_csv('mae_greens_rfr.csv')
mae_reds_rfr = pd.read_csv('mae_reds_rfr.csv')
```

```python
mae_global_rfr_wo_static = pd.read_csv('mae_global_rfr_wo_static.csv')
mae_greens_rfr_wo_static = pd.read_csv('mae_greens_rfr_wo_static.csv')
mae_reds_rfr_wo_static = pd.read_csv('mae_reds_rfr_wo_static.csv')
```

```python
# Global
fig = go.Figure()

# CRTM API

fig.add_trace(
    go.Scatter(x=mae_global_crtm['remaining_seconds'], y=mae_global_crtm['mae'], name='CRTM API'))

# MT

fig.add_trace(
    go.Scatter(x=mae_global_mt['remaining_seconds'], y=mae_global_mt['mae'], name='Model Tree'))


# RFR

fig.add_trace(
    go.Scatter(x=mae_global_rfr['remaining_seconds'], y=mae_global_rfr['mae'], name='Random Forest'))


# RFR w/o static

fig.add_trace(
    go.Scatter(x=mae_global_rfr_wo_static['remaining_seconds'],
               y=mae_global_rfr_wo_static['mae'],
               name='Random Forest w/o static'))


fig.update_xaxes(autorange="reversed")

# Set y-axes titles
fig.update_yaxes(title_text="<b>MAE</b> (minutes)")

# Edit the layout
# title='Global estimation error per minute'
fig.update_layout(xaxis_title='Remaining time (minutes)')

fig.show()
```

```python
# Real-time
fig = go.Figure()

# CRTM API

fig.add_trace(
    go.Scatter(x=mae_greens_crtm['remaining_seconds'], y=mae_greens_crtm['mae'], name='CRTM API'))

# MT

fig.add_trace(
    go.Scatter(x=mae_greens_mt['remaining_seconds'], y=mae_greens_mt['mae'], name='Model Tree'))


# RFR

fig.add_trace(
    go.Scatter(x=mae_greens_rfr['remaining_seconds'], y=mae_greens_rfr['mae'], name='Random Forest'))


# RFR w/o static

fig.add_trace(
    go.Scatter(x=mae_greens_rfr_wo_static['remaining_seconds'],
               y=mae_greens_rfr_wo_static['mae'],
               name='Random Forest w/o static'))


fig.update_xaxes(autorange="reversed")

# Set y-axes titles
fig.update_yaxes(title_text="<b>MAE</b> (minutes)")

# Edit the layout
# title='Real-time based observations estimation error per minute'
fig.update_layout(xaxis_title='Remaining time (minutes)')

fig.show()
```

```python
# Non real-time
fig = go.Figure()

# CRTM API

fig.add_trace(
    go.Scatter(x=mae_reds_crtm['remaining_seconds'], y=mae_reds_crtm['mae'], name='CRTM API'))

# MT

fig.add_trace(
    go.Scatter(x=mae_reds_mt['remaining_seconds'], y=mae_reds_mt['mae'], name='Model Tree'))

# RFR

fig.add_trace(
    go.Scatter(x=mae_reds_rfr['remaining_seconds'], y=mae_reds_rfr['mae'], name='Random Forest'))

# RFR w/o static

fig.add_trace(
    go.Scatter(x=mae_reds_rfr_wo_static['remaining_seconds'],
               y=mae_reds_rfr_wo_static['mae'],
               name='Random Forest w/o static'))


fig.update_xaxes(autorange="reversed")

# Set y-axes titles
fig.update_yaxes(title_text="<b>MAE</b> (minutes)")

# Edit the layout
# title='Non real-time based observations estimation error per minute'
fig.update_layout(xaxis_title='Remaining time (minutes)')

fig.show()
```
